```python
import numpy as np
import pandas as pd
import json
import plotly.express as px
```

```python
import pickle
```

```python
df = pd.read_csv("data/map_data_final.csv")
df.head()
```

```python
all_sect = json.load(open("geojson/sectors.geojson","r"))
```

```python
for feature in all_sect["features"]:
    if 'Sector' not in feature['properties']:
        print("Feature without 'Sector':", feature)
```

```python
id_map = []
for feature in all_sect["features"]:
        feature['id'] = feature['properties']['Sector']
        id_map.append(feature['id'])
```

```python
fig = px.choropleth_mapbox(df[df['property_type'] == "Flat/Apartment"],locations="Sector",geojson=all_sect,color="price_per_sqft",
                           mapbox_style="open-street-map",color_continuous_scale="viridis",center = {"lat":30.689281,"lon":76.786950},zoom = 10)
fig.show()
```
