```python
import numpy as np
import pandas as pd
import json
import plotly.express as px
```

```python
df = pd.read_csv("data/final_df_v2.csv")
df.head()
```

```python
df['price_per_sqft'] = (df["price"] * 10000000)/df["Area"]
```

```python
temp_df = df[['City','property_type','Sector','price_per_sqft']].copy()
```

```python
temp_df
```

For convenience we will change builder floor to flat/apartment

```python
temp_df['property_type'].value_counts()
```

```python
temp_df.replace("Builder Floor","Flat/Apartment",inplace = True)
```

```python
temp_df['property_type'].value_counts()
```

Lets groupby on basis of city and property_type

```python
group = temp_df.groupby(["property_type","City","Sector"])
```

```python
group.size()
```

```python
final_df = group['price_per_sqft'].mean().reset_index()
```

```python
pd.options.display.max_rows = None
```

```python
final_df["price_per_sqft"] = round(final_df["price_per_sqft"],2)
```
