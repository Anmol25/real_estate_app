```python
all_present_df = df[((~df['Super Built Up'].isna()) & (~df['Built Up area'].isna()) & (~df['Carpet area'].isna()))]
```

```python
super_area_ratio = (all_present_df['Super Built Up'] / all_present_df['Built Up area']).median()
```

```python
carpet_area_ratio = (all_present_df['Carpet area'] / all_present_df['Built Up area']).median()
```

```python
print(f"Super Built Up area ratio : {super_area_ratio}")
print(f"Carpet area ratio : {carpet_area_ratio}")
```

```python
avl_s_c = df[((~df['Super Built Up'].isna()) & (df['Built Up area'].isna()) & (~df['Carpet area'].isna()))]
```

```python
avl_s_c['Built Up area'].fillna(round(((avl_s_c['Super Built Up']/1.182) + (avl_s_c['Carpet area']/0.83))/2),inplace = True)
```

```python
df.update(avl_s_c)
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
avl_s = df[((~df['Super Built Up'].isna()) & (df['Built Up area'].isna()) & (df['Carpet area'].isna()))]
```

```python
avl_s['Built Up area'].fillna(round((avl_s['Super Built Up']/1.182)),inplace = True)
```

```python
df.update(avl_s)
```

```python
# Only Carpet area available
avl_c = df[((df['Super Built Up'].isna()) & (df['Built Up area'].isna()) & (~df['Carpet area'].isna()))]
```
