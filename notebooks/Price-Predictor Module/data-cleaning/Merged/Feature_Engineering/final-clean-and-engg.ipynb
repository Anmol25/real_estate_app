```python
df['TotalFloors'].value_counts().sort_index()
```

```python
def categorize_floor(floor):
    if floor <= 3:
        return 'Low-Rise'
    elif 4 <= floor <= 12:
        return 'Mid-Rise'
    elif 13 <= floor <= 40:
        return 'High-Rise'
    else:
        return 'Ultra-High-Rise'
```

```python
df['TotalFloors'].apply(categorize_floor)
```

```python
df.insert(27,'FloorRise',value = df['TotalFloors'].apply(categorize_floor))
```
