### Floors

```python
df['noOfFloor'].value_counts()
```

```python
df['noOfFloor'].value_counts()
```

```python
def floor_func(x):
    floor_no = 0
    Total_floor = 0
    if pd.notna(x):
        temp = x.split(' ')
        if temp[0][0].isnumeric():
            floor_no = re.findall(r'\d+', temp[0])[0]
        else:
            if temp[0] == 'Ground' or temp[0] == 'Lower':
                floor_no = 0
            elif temp[0] == 'Basement':
                floor_no = -1
            else:
                floor_no = -2
        Total_floor = temp[-2]
    return floor_no,Total_floor
```

```python
df['noOfFloor'].apply(floor_func)
```

```python
floors_df = df['noOfFloor'].apply(floor_func).apply(pd.Series)
floors_df.columns = ['FloorNo', 'TotalFloors']
floors_df
```

```python
df.insert(15,'TotalFloors',value= floors_df['TotalFloors'].values)
```

```python
df.insert(15,'FloorNo',value= floors_df['FloorNo'].values)
```

```python
df['TotalFloors'] = df['TotalFloors'].astype(int)
df['FloorNo'] = df['FloorNo'].astype(int)
```

```python
df.drop(columns = 'noOfFloor',inplace = True)
```

```python
def price_func(x):
    if pd.notna(x):
        temp = x.split(' ')
        if temp[-1] == 'Cr':
            return float(temp[0])
        elif temp[-1] == 'Lac':
            return (float(temp[0]))/100
        else:
            return x
    else:
        return x
```

### Parking

```python
df['Parking']
```

```python
def park_func(x):
    Covered = 0
    Open = 0
    if pd.notna(x):
        temp = x.split(',')
        if len(temp) >= 2:
            # Covered
            cov_tmp = temp[0].strip().split(' ')
            if cov_tmp[0].isnumeric():
                Covered = int(cov_tmp[0])
                
            # Open
            open_tmp = temp[-1].strip().split(' ')
            if open_tmp[0].isnumeric():
                Open = int(open_tmp[0])
        else:
            if temp[0].split(' ')[0].isnumeric():
                if 'Covered' in temp[0]:
                    Covered = temp[0].split(' ')[0]
                else:
                    Open = temp[0].split(' ')[0]
    return Covered,Open    
```

```python
df['Parking'].apply(park_func)
```

```python
parking_df = df['Parking'].apply(park_func).apply(pd.Series)
parking_df.columns = ['CoveredParking', 'OpenParking']
parking_df
```

```python
df.insert(22,'OpenParking',value= parking_df['OpenParking'].values)
```

```python
df.insert(22,'CoveredParking',value= parking_df['CoveredParking'].values)
```

```python
df.drop(columns = 'Parking',inplace = True)
```
