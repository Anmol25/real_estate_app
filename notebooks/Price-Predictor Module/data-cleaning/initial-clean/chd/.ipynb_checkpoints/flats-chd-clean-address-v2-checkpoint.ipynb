```python
def address_func(x):
    if pd.notna(x):
        temp =  x.split(',')
        return temp[-3].strip(),temp[-2].strip()
    else:
        return x
```

```python
df['address1'].apply(address_func)
```

```python
df.insert(5,'City',value = temp_df['City'].values)
df.insert(5,'Sector',value = temp_df['Sector'].values)
```

```python
def sector_filt(sector):
    return df[df['Sector'] == sector][['Society','Sector','City']]
```

```python
# 72
df['Sector'].value_counts()
```

```python
def sector_clean(x):
    if pd.notna(x):
        if x.startswith('Sector'):
            output_str = re.sub(r'(\d+)([A-Za-z])', r'\1', x)
            return output_str
        elif x[0].isdigit():
            output_str = re.sub(r'\d+[A-Za-z]', '', x, count=1)
            return output_str.strip()
        else:
            return x
    else:
        return x
```

```python
def sect_corrector(sector_to_correct, new_sec_val = '', new_city_val = '', new_society_val = '', sector = False,city = False, society = False):
    temp_rows = sector_filt(sector_to_correct).index
    for i in temp_rows:
        if sector == True:
            df['Sector'].loc[i] = new_sec_val
        if city == True:
            df['City'].loc[i] = new_city_val
        if society == True:
            df['Society'].loc[i] = new_society_val
```
