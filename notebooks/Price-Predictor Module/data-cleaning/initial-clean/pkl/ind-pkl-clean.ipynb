```python
def price_sq_func(x):
    if pd.notna(x):
        numbers = re.findall(r'\d+', x)
        return float(''.join(numbers))
    else:
        return x
```

```python
def extract_numbers(input_string):
    # Use regular expression to find all numbers
    numbers = re.findall(r'\d+', input_string)
    # Join the found numbers into a single string
    return float(''.join(numbers))
```

```python
extract_numbers('2,90,000 per sq.yards')
```

```python
def address_func(x):
    if pd.notna(x):
        temp =  x.split(',')
        try:
            Society = temp[-4].strip()
        except:
            Society = ''
        return Society,temp[-3].strip(),temp[-2].strip()
    else:
        return x
```
